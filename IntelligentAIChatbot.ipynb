```python
# Question and Answering
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pyttsx3
import speech_recognition as sr
import os

def Jacard_Similarity(text):
    if 'Question' not in dataset.columns or 'Answer' not in dataset.columns:
        raise ValueError("The dataset must contain both 'Question' and 'Answer' columns.")
    # Handle missing values by replacing np.nan with an empty string
    dataset['Question'] = dataset['Question'].fillna('')
    # Calculate TF-IDF vectors for the text and the 'Question' column
    vectorizer = TfidfVectorizer()
    text_vector = vectorizer.fit_transform([text])
    question_vectors = vectorizer.transform(dataset['Question'])
    # Calculate cosine similarity
    similarities = cosine_similarity(text_vector, question_vectors).flatten()
    # Find the index of the most similar question
    most_similar_index = similarities.argmax()
    # Get the most similar question and its corresponding answer
    most_similar_question = dataset.loc[most_similar_index, 'Question']
    most_similar_answer = dataset.loc[most_similar_index, 'Answer']

    return most_similar_question, most_similar_answer

# Speech to Text function
def speech_to_text(audio_file_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio_file_path) as source:
            audio = recognizer.record(source)
        # Perform speech recognition
        text = recognizer.recognize_google(audio)
        return text
    except Exception as e:
        print(f"Error converting speech to text: {e}")
        return None

# Text to Speech function
def text_to_speech(text):
    output_file_path = "/content/drive/MyDrive/NLP_PROJECT/output.wav"
    engine = pyttsx3.init()
    engine.save_to_file(text, output_file_path)
    engine.runAndWait()
```

```python
# QnA_Menu
def QnA_Menu():
    while True:
        print("\033[1m==============================================================================================\033[0m")
        print("\033[1m\t \t \t ********** Question and Answering **********\033[0m")
        print("0. Stop")
        print("1. Text to Speech")
        print("2. Speech to Text Q&A")
        print("3. Text to Speech Q&A")
        choice = input("Enter  choice : ")
        print("\n")
        if choice == "0":
            print("\t\t\tTerminated\n")
            print("\033[1m==============================================================================================\033[0m")
            break
        elif choice == "1":
            text = input("Enter the text you want to convert to speech: ")
            text_to_speech(text)
            print("Text is converted to speech \n")
            print("\033[1m==============================================================================================\033[0m")
        elif choice == "2":
            # Speech to Text for the question
            print("Auto Question Files [Q1,Q2,Q3,Q4,Q5]")
            file_name = input("Enter file for (without extension): ")
            audio_file_path = f"/content/drive/MyDrive/NLP_PROJECT/{file_name}.wav"
            print(f"Attempting to open file: {audio_file_path}")
            user_input = speech_to_text(audio_file_path)
            print("Speech to Text :",user_input)
            if user_input:
                # Find matching question in the dataset
                Que,Ans=Jacard_Similarity(user_input)
                print("Question:", Que)
                print("Answer:", Ans)
            else:
                print("Error in speech recognition. Please try again.")
            print("\033[1m==============================================================================================\033[0m")
        elif choice=="3":
            text = input("Enter Question : ")
            # Find matching question in the dataset
            Que,Ans=Jacard_Similarity(text)
            print("Question:", Que)
            print("Answer:", Ans)
            text_to_speech(text)
            print("Answer is saved in speech")
            print("\033[1m==============================================================================================\033[0m")
        else:
            print("Invalid choice.")
            print("\033[1m==============================================================================================\033[0m")
```

```python
#Running QnA Module
QnA_Menu()
```

```python
# Sentimental Analysis
from transformers import pipeline
import pyttsx3
import speech_recognition as sr
from pydub import AudioSegment

# Perform Sentiment Analysis function
def perform_sentiment_analysis(sentiment_pipeline, input_text):
    results = sentiment_pipeline(input_text)
    sentiment_label = results[0]['label']
    sentiment_score = results[0]['score']

    # Text to Speech
    output_text = f"Your sentence is {sentiment_label} with a confidence score of {sentiment_score:.2f}."
    text_to_speech(output_text)
    print(output_text)

# Speech to Text function
def speech_to_text(audio_file_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio_file_path) as source:
            audio = recognizer.record(source)
        # Perform speech recognition
        text = recognizer.recognize_google(audio)
        print("Transcription:", text)
        return text
    except Exception as e:
        print(f"Error converting speech to text: {e}")
        return None

# Text to Speech function
def text_to_speech(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()
```

```python
# Sentiment Analysis Menu
def SentimentalAnalysis_Menu():
    try:
        # Load sentiment analysis model
        sentiment_pipeline = pipeline("sentiment-analysis")
    except Exception as e:
        print(f"Error loading sentiment analysis model: {e}")
        return

    while True:
        print("\033[1m==============================================================================================\033[0m")
        print("\033[1m\t \t \t ********** Sentiment Analysis **********\033[0m")
        print("0. Stop")
        print("1. Perform Text-based Sentiment Analysis")
        print("2. Perform Voice-based Sentiment Analysis")
        choice = input("Enter choice : ")
        if choice == "0":
            print("Terminated.\n")
            print("\033[1m==============================================================================================\033[0m")
            break
        elif choice == "1":
            text_input = input("Enter the text for sentiment analysis: ")
            perform_sentiment_analysis(sentiment_pipeline, text_input)
            print("\033[1m==============================================================================================\033[0m")
        elif choice == "2":
            print("Audio Files (S1,S2,S3,S4,S5)")
            file_name = input("Enter the name of file (without extension): ")
            audio_file_path = f"/content/drive/MyDrive/NLP_PROJECT/{file_name}.wav"
            user_input = speech_to_text(audio_file_path)
            if user_input:
                perform_sentiment_analysis(sentiment_pipeline, user_input)
            print("\033[1m==============================================================================================\033[0m")
        else:
            print("Invalid choice")
            print("\033[1m==============================================================================================\033[0m")

```

```python
# Running Sentimental Analysis
SentimentalAnalysis_Menu()
```

```python
#Translation
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
def Translator(Text):
    # Load tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained("ZainAli60/English-to-Urdu")
    model = AutoModelForSeq2SeqLM.from_pretrained("ZainAli60/English-to-Urdu")
    # Tokenize input text
    input_ids = tokenizer(Text, return_tensors="pt").input_ids
    # Generate translation
    translation = model.generate(input_ids)
    # Decode the generated output
    TranslatedText = tokenizer.decode(translation[0], skip_special_tokens=True)
    return TranslatedText

# Speech to Text function
def speech_to_text(audio_file_path):
    recognizer = sr.Recognizer()
    try:
        with sr.AudioFile(audio_file_path) as source:
            audio = recognizer.record(source)
        # Perform speech recognition
        text = recognizer.recognize_google(audio)
        print("Transcription:", text)
        return text
    except Exception as e:
        print(f"Error converting speech to text: {e}")
        return None

# Text to Speech function
def text_to_speech(text):
    output_file_path = "/content/drive/MyDrive/NLP_PROJECT/output.wav"
    engine = pyttsx3.init()
    engine.save_to_file(text, output_file_path)
    engine.runAndWait()
```

```python
# Translation_Menu
def Translation_Menu():
    while True:
        print("\033[1m==============================================================================================\033[0m")
        print("\033[1m\t \t \t ********** Translation **********\033[0m")
        print("0_ Stop")
        print("1_ Text Translation ")
        print("2_ Voice  Translation ")
        choice = input("Choose Option: ")
        if choice == "0":
            print("Terminated")
            print("\033[1m==============================================================================================\033[0m")
            break
        elif choice == "1":
            Text = input("Enter text in English: ")
            TranslatedText = Translator(Text)
            print("Original Text:", Text)
            print("Translated Text:", TranslatedText)
            print("\033[1m==============================================================================================\033[0m")
        elif choice =="2":
            print("Audio Files [T1,T2,T3,T4,T5]")
            file_name = input("Enter file name (without extension): ")
            audio_file_path = f"/content/drive/MyDrive/NLP_PROJECT/{file_name}.wav"
            Original = speech_to_text(audio_file_path)
            Translated = Translator(Original)
            print("Original Sentence:", Original)
            print("Translated Sentence:", Translated)
            text_to_speech(Translated)
            print("\033[1m==============================================================================================\033[0m")
        else:
            print("Wrong Choice. Enter Again\n")
            print("\033[1m==============================================================================================\033[0m")
```

```python
# Running Translation Module
Translation_Menu()
```
